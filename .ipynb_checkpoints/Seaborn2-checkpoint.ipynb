```python
import seaborn as sns
```

## Categorical Plots

### Categorical Scatter Plot

gives Bivariate analysis
- Stripplot
- Swarmplot

### Categorical Distribution Plots

give univariate analysis
- Boxplot
- Violinplot

### Categorical Estimate Plot -> for central tendency

- Barplot
- Pointplot

### Figure level function -> `catplot`

```python
tips
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
# strip plot
# works good on medium and high size data
# axes level function
sns.stripplot(data=tips,x='day',y='total_bill',color='red')
```

```python
sns.stripplot(data=tips,x='day',y='total_bill',color='red',jitter=False)
```

```python
# using cat plot
sns.catplot(data=tips,y='total_bill',x='day',kind='strip',hue='day')
```

```python
# jitter
sns.catplot(data=tips,y='total_bill',x='day',kind='strip',hue='day',jitter=.05)
```

```python
# swarmplot
sns.catplot(data=tips,y='total_bill',x='day',kind='swarm',hue='day')
```

```python
#axis level
sns.swarmplot(data=tips,y='total_bill',x='day',hue='day')
```

```python
# hue
# small insect
# it work good on the small data
sns.swarmplot(data=tips,y='total_bill',x='day',hue='sex')
```

### Boxplot

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (â€œminimumâ€, first quartile [Q1], median, third quartile [Q3] and â€œmaximumâ€). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

image.png

```python
# Box plot is a axis level function
sns.boxplot(data=tips,x='day',y='total_bill',hue='day')
```

```python
sns.boxplot(data=tips,x='sex',y='total_bill',hue='sex')
```

```python
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex')
```

```python
# catplot - is a figure level function
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',kind='box')
```

```python
# single boxplot -> numerical col
sns.boxplot(data=tips,y='total_bill')
```

```python
# Countplot
sns.countplot(data=tips,x='sex')
```

```python
sns.catplot(data=tips,x='sex',y='total_bill',col='smoker',kind='box',hue='sex',row='time')
```

### Regression Plots

- regplot
- lmplot

In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

```python
# axes level
# reg plot
# hue is not available
sns.regplot(data=tips,x='total_bill',y='tip')
```

```python
# lmplot
# in lmplot and regplot only difference is  we add hue parametw=er in lmplot
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
# residplot
# error in prediction
sns.residplot(data=tips,x='total_bill',y='tip')
```

### A second way to plot Facet plots -> FacetGrid

- figure level -> relplot -> displot -> catplot -> lmplot


```python
sns.catplot(data=tips,y='total_bill',x='sex',kind='violin',hue='sex',col='day',row='time')
```

```python
g = sns.FacetGrid(data=tips,col='day',row='time',hue='smoker')
```

```python
g = sns.FacetGrid(data=tips,col='day',row='time',hue='smoker')
g.map(sns.boxplot,'tip','total_bill')
g.add_legend()
```

### Plotting Pairwise Relationship (PairGrid Vs Pairplot)

```python
iris= sns.load_dataset('iris')
iris
```

```python
sns.pairplot(iris,hue='species')
```

```python
# pair grid
g = sns.PairGrid(data=iris)
# g map
g.map(sns.scatterplot)
```

```python
# map_diag -> map_offdiag
sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
# map_diag -> map_offdiag
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.kdeplot)
```

```python
# map_diag -> map_upper -> map_lower
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
# vars
g = sns.PairGrid(data=iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```
