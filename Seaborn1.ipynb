### 1. Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# relplot -> figure level -> square shape
tips= sns.load_dataset('tips')
tips
```

```python
# scatter plot-> axes level function
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
# relplot -> figure level -> squareplot
sns.relplot(data= tips,x='total_bill',y = 'tip',kind = 'scatter')
```

```python
# relplot -> figure level -> squareplot
sns.relplot(data= tips,x='total_bill',y = 'tip',kind = "line")
```

# HUE

```python
# relplot -> figure level -> squareplot
sns.relplot(data= tips,x='total_bill',y = 'tip',kind = "scatter",hue="sex")
```

# Style

```python
# relplot -> figure level -> squareplot
sns.relplot(data= tips,x='total_bill',y = 'tip',kind = "scatter",hue="sex",style='time')
```

```python
# scatter using relplot -> size and hue,style
sns.relplot(data= tips,x='total_bill',y = 'tip',kind = "scatter",hue="sex",style='time',size='size')
```

2. Matrix
    - Heatmap
    - Clustermap

```python
gap = px.data.gapminder()
gap
```

```python
# heatmap
import pandas as pd
df = pd.pivot_table(data=gap,index='country',columns='year',values='lifeExp')
```

```python
df
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.xlabel('year')
plt.imshow(df,cmap='viridis',aspect='auto')
plt.colorbar()
```

```python
temp_df = gap.pivot(index='country',columns='year',values='lifeExp')
plt.figure(figsize=(15,15))
sns.heatmap(temp_df,cmap='viridis')
```

```python
df = gap[gap['continent']=='Europe']
df
df=df.pivot(columns='year',index='country',values='lifeExp')
plt.figure(figsize=(15,15))
sns.heatmap(df,cmap='viridis',annot=True)
```

```python
df = gap[gap['continent']=='Asia']
df
df=df.pivot(columns='year',index='country',values='lifeExp')
plt.figure(figsize=(15,15))
sns.heatmap(df,cmap='summer',annot=True,linewidth=0.5)
```
