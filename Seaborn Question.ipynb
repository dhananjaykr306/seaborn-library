    @Title : 1. Seaborn Bar plot  Write a program to draw bar plot of sex against survived for a dataset given in the url https://github.com/mwaskom/seaborn-data/blob/master/titanic.csv 


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
titanic = sns.load_dataset('titanic')
```

```python
titanic
```

```python
sns.barplot(data=titanic,x='sex',y='survived',hue='sex')
```

    @Title : Seaborn pointplot 1. Write a program to draw a point plot for sex against survived for a dataset given in url https://github.com/mwaskom/seaborn-data/blob/master/titanic.csv 


```python
import numpy as np
sns.pointplot(data=titanic, x='sex', y='survived', hue='sex',estimator='mean', palette='deep',legend=False)
plt.title('Survival Rate by Sex on the Titanic', fontsize=16)
plt.xlabel('Sex', fontsize=12)
plt.ylabel('Survival Rate', fontsize=12)
plt.show()
```
